### 1.3. Preprocessing Awal Untuk Data Yang Sudah Di Crawling

#### Menghilangkan Karakter tidak penting pada Isi Tweet Yang telah di Crawling/Noise Reduction 

```python
# Menghilangkan tweet dan username dari kalimat-kalimat, simbol-simbol, emoticon yang tidak dibutuhkan selama pengujian menggunakan regex (regular expression)
new_data_tweet = data_tweet.copy()
import re
def bersih_text(text):
    #menghilangkan link tweet
    text = re.sub(r"https:(\/\/t\.co\/([A-Za-z0-9]|[A-Za-z]){10})", " ", text)
    #menghilangkan tag username pada tweet
    text = re.sub("(@[A-Za-z0-9_]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",text)
    #menghilangkan kalimat, simbol, emoticon yang tidak dibutuhkan pada sentiment analysis
    text = re.sub(r'[^a-zA-z0-9]'," ", text)
    text = re.sub(r'\b\w{1,2}\b'," ", text)
    text = re.sub(r'\s\s+'," ", text)
    return text

new_data_tweet["Tweet"] = data_tweet["Tweet"].apply(bersih_text)

new_data_tweet
```

#### Case Folding 

```python
# Mengubah setiap kalimat yang ada di kolom "Username" dan "Tweet" menjadi lower case/huruf kecil
new_data_tweet["Username"] = new_data_tweet["Username"].str.lower()
new_data_tweet["Tweet"] = new_data_tweet["Tweet"].str.lower()

new_data_tweet.head()
```

### 2.3 Tokenizing

Tokenizing/Tokenisasi adalah tahapan awal dalam text-preprocessing untuk memisahkan potongan kata dari tiap kalimat, paragraf menjadi perkata(token)

```python
#Melakukan tokenizing
df_tweet_clean_text = []
for i in range(len(df_tweet)):
    x = df_tweet["Tweet"][i]
    x_word_tokens = word_tokenize(x)

    # menghilangkan tanda baca
    x_word_tokens_removed_punctuations = [w for w in x_word_tokens if w not in punctuation]

    # menghilangkan angka/numeric
    x_word_tokens_removed_punctuations = [w for w in x_word_tokens_removed_punctuations if w.isalpha()]
    
    df_tweet_clean_text.append(x_word_tokens_removed_punctuations)


df_tweet["tweet_tokenizing"] = df_tweet_clean_text
df_tweet
```

### 2.5 Stemming

```python
# Membuat Stemmer dari sastrawi
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Membuat fungsi stemming
def stemming(text):
    return stemmer.stem(text)

term_dict = {}

# Menghitung ada berapa kata/words di dalam kolom "Tweet_Normalisasi"
for document in df_tweet["Tweet_Stopwords"]:
    for text in document:
        if text not in term_dict:
            term_dict[text] = " "
            
print(len(term_dict))
print("------------------------")

# Menampilkan isi kata/words yang telah dilakukan proses stemming
for text in term_dict:
    term_dict[text] = stemming(text)
    print(text,":" ,term_dict[text])
    
print(term_dict)
print("------------------------")

# Membuat fungsi baru untuk proses stemming
def stemming_word(document):
    return [term_dict[text] for text in document]

df_tweet["Tweet_Stemming"] = df_tweet["Tweet_Stopwords"].swifter.apply(stemming_word)
```

###  Labelling menggunakan TextBlob

```python
def getSubjectivity(review):
  analysis = TextBlob(review)
  try:
    return analysis.translate(from_lang="id", to='en').sentiment.subjectivity
  except:
    return analysis.sentiment.subjectivity

def getPolarity(review):
  analysis = TextBlob(review)
  try:
    return analysis.translate(from_lang="id", to="en").sentiment.polarity
  except:
    return analysis.sentiment.polarity

def analyze(score):
  if score < 0:
    return 'Negative'
  elif score == 0:
    return 'Neutral'
  else:
    return 'Positive'
```

```python
#final_data["Subjectivity"] = final_data["Tweet_Bersih"].apply(getSubjectivity)
df_tweet["Polarity"] = df_tweet["Tweet_Stemming"].apply(getPolarity)
df_tweet["Sentimen"] = df_tweet["Polarity"].apply(analyze)
df_tweet
```
